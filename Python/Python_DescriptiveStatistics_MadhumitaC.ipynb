# 3. The following table shows the mean and variance of the expenditure for two groups of people. You want to compare the variability in expenditure for both groups with respect to their mean. Which statistical measure would you use to evaluate the variability in expenditure? Please provide an explanation for your answer.

Since the means of the two groups (500,000 and 40,000) are quite different, comparing the standard deviations directly would not give a clear picture of relative variability.
The CV(Coefficient of Variation) allows us to understand how much variability exists in relation to the average expenditure, which helps in making more meaningful comparisons between the groups.
CV=(Standard Deviation/Mean)×100
This measure provides a standardized way to compare variability across different datasets, even when the means are significantly different.

```python
# Given data
mean_group1 = 500000
std_dev_group1 = 125000

mean_group2 = 40000
std_dev_group2 = 10000

# Function to calculate Coefficient of Variation
def calculate_cv(mean, std_dev):
    return (std_dev / mean) * 100

# Calculating CV for both groups
cv_group1 = calculate_cv(mean_group1, std_dev_group1)
cv_group2 = calculate_cv(mean_group2, std_dev_group2)

# Output the results
print(f"Coefficient of Variation for Group 1: {cv_group1:.2f}%")
print(f"Coefficient of Variation for Group 2: {cv_group2:.2f}%")
```

Insights:

Both groups have the same CV, indicating that their variability in expenditure is proportionate to their respective means. This means that, in terms of relative variability, the two groups experience similar levels of dispersion in their spending.

A CV of 25% implies that the standard deviation is 25% of the mean for both groups. This suggests that in both groups, expenditures are consistently varied around their respective means, which can be interpreted as a moderate level of risk in spending behavior.


Conclusion:
In summary, while both groups display a similar level of relative variability in their expenditures as indicated by the Coefficient of Variation, the differences in their absolute expenditure amounts remain significant. This finding can inform targeted approaches in marketing, policy, or other decision-making processes, emphasizing the need to balance both relative and absolute figures in analysis.

# 5. Assume you are the trader and you have invested over the years, and you are worried about the average return on investment. What average method would you use to compute the average return for the data given below?

```python
# assign data
myinvest = [
    ["2015", 36,5000], 
    ["2016", 24,6400], 
    ["2017", -48,7890], 
    ["2018", -30,9023],
    ["2019", 15,4567],
    ["2020", 31,3890],
]
 
# create header
head = ["Year", "Return %","Asset Price"]
 
# display table
print(tabulate(myinvest, headers=head, tablefmt="grid"))
```

To compute the average return on investment based on the provided data, you can use the weighted average method, which takes into account the percentage returns for each year while considering the actual performance.

Weighted Contributions: Since the asset price varies each year, using the weighted average allows you to account for the size of the investment in each year. This way, years with higher asset prices (and thus higher potential impact) have a greater influence on the average return.

Realistic Representation: The weighted average provides a more realistic view of your overall performance since it reflects the actual returns in relation to the size of the investment in each period.

```python
# Calculate weighted average return
total_weighted_return = 0
total_asset_prices = 0

for year_data in myinvest:
    return_percentage = year_data[1]
    asset_price = year_data[2]
    total_weighted_return += return_percentage * asset_price
    total_asset_prices += asset_price
#print(total_weighted_return)
#print(total_asset_prices)
# Weighted average return calculation
weighted_average_return = total_weighted_return / total_asset_prices

print(f"\nWeighted Average Return: {weighted_average_return:.2f}%")
```

Using the weighted average method provides a more accurate reflection of your overall return on investment over the years, considering the impact of both return percentages and asset prices.

Statistic: A statistic, on the other hand, describes a characteristic of a sample drawn from the population. The average height calculated from your sample would be a statistic.

# 12. What is the percentage of variation you can observe in the ‘Income’ variable?

# 20. Using the z-score of the income variable, find out the number of observations outside the +-3σ.

```python
# Calculate the mean and standard deviation of the 'Income' variable
mean_income = df_bank['Income'].mean()
std_income = df_bank['Income'].std()

# Calculate the z-scores
df_bank['z_score_income'] = (df_bank['Income'] - mean_income) / std_income

# Count the number of observations outside ±3σ
outliers_count = ((df_bank['z_score_income'] < -3) | (df_bank['z_score_income'] > 3)).sum()

print(f'Number of observations outside ±3σ: {outliers_count}')
```

If your analysis indicates that there are 2 observations outside ±3σ for the 'Income' variable, it means that these two data points are considered outliers.
These are extreme values that may indicate unusual income levels.
