```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("flights.csv")
print(df.head(10))

plt.figure(figsize=(8,5))
sns.histplot(df['passengers'], kde=True)
#plt.grid(True)
plt.title("passengers Count")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(df['passengers'])
#plt.grid(True)
plt.title("passengers distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x=df['month'], y=df['passengers'])
#plt.grid(True)
plt.title(" month / passenger distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=df['month'], y=df['passengers'])
#plt.grid(True)
plt.title(" month / passenger distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x=df['year'], y=df['passengers'])
#plt.grid(True)
plt.title(" year / passenger distribution")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x=df['year'], y=df['passengers'])
plt.title(" year / passenger distribution")
plt.show()
```
